```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder



# -----------------------------

# 1) 데이터 준비

# -----------------------------
cols = ["price", "maint", "doors", "persons",	"lug_capacity",	"safety",	"output"]

df = pd.read_csv('/content/drive/MyDrive/car_evaluation.csv', header=None, names=cols).dropna()  # CSV 파일 경로

label_encoders = {}
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

df
```

```python
X = df.drop(columns=["output"])  # output 버림

y = df["output"] # output 얻음
```

```python
X_train, X_test, y_train, y_test = train_test_split(  # train : 정답값

    X, y, test_size=0.2, random_state=42  # stratify : y값 레이블을 비슷하게 맞춤

)
```

```python
# -----------------------------

# 2) 모델 구성

# -----------------------------

dt = DecisionTreeClassifier(random_state=42)  # 훈련 모델 만들기

rf = RandomForestClassifier(n_estimators=200, random_state=42)

lr = LogisticRegression(max_iter=500)
```

```python
# -----------------------------

# 3) 모델 학습

# -----------------------------

dt.fit(X_train, y_train)

rf.fit(X_train, y_train)

lr.fit(X_train, y_train)
```

```python
# -----------------------------

# 4) 모델 평가

# -----------------------------

dt_acc = accuracy_score(y_test, dt.predict(X_test)) # predict 실제 예측할만 할만한 데이터

rf_acc = accuracy_score(y_test, rf.predict(X_test))

lr_acc = accuracy_score(y_test, lr.predict(X_test))
```

```python

print("=== Test Accuracy ===")

print(f"Decision Tree : {dt_acc:.4f}")

print(f"Random Forest : {rf_acc:.4f}")

print(f"Logistic Reg. : {lr_acc:.4f}")
```
